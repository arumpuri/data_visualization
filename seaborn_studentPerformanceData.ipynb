```python
# line plot
sns.lineplot(data=student_data, x="Attendance", y="Exam_Score", hue="Gender")

# Add title and labels
plt.title("How Attendance Affects Exam Scores")
plt.xlabel("Attendance (days)")
plt.ylabel("Exam Score")

# Add background
sns.set_style("whitegrid")

plt.show()
```

```python
# line plot
sns.lineplot(data=student_data, x="Attendance", y="Exam_Score", hue="Gender", palette="coolwarm")

# Add title and labels
plt.title("How Attendance Affects Exam Scores")
plt.xlabel("Attendance (days)")
plt.ylabel("Exam Score")

# Add background
sns.set_style("whitegrid")

plt.show()
```

```python
# line plot
custom_palette = sns.color_palette(["#FF5733", "#33FF57"])  # Hex colors
sns.lineplot(data=student_data, x="Attendance", y="Exam_Score", hue="Gender", palette=custom_palette)


# Add title and labels
plt.title("How Attendance Affects Exam Scores")
plt.xlabel("Attendance (days)")
plt.ylabel("Exam Score")

# Add background
sns.set_style("whitegrid")

plt.show()
```

```python
 # line plot
# Add error bars (standard deviation)
# Differentiate line styles by gender
custom_palette = sns.color_palette(["#FF5733", "#33FF57"])  # Hex colors
sns.lineplot(data=student_data, x="Attendance", y="Exam_Score", hue="Gender", palette=custom_palette, errorbar="sd", style="Gender")


# Add title and labels
plt.title("How Attendance Affects Exam Scores")
plt.xlabel("Attendance (days)")
plt.ylabel("Exam Score")

# Add background
sns.set_style("whitegrid")

plt.show()
```

```python
 # line plot
# Add error bars (standard deviation)
# Differentiate line styles by gender# Apply custom dashes for different categories
custom_palette = sns.color_palette(["#FF5733", "#33FF57"])  # Hex colors
sns.lineplot(data=student_data, x="Attendance", y="Exam_Score", hue="Gender", palette=custom_palette, errorbar="sd", style="Gender", dashes=[(2, 2), (4, 4)])


# Add title and labels
plt.title("How Attendance Affects Exam Scores")
plt.xlabel("Attendance (days)")
plt.ylabel("Exam Score")

# Add background
sns.set_style("whitegrid")

plt.show()
```

```python
 # line plot
# Add error bars (standard deviation)
# Differentiate line styles by gender# Apply custom dashes for different categories# Add markers to the plot

custom_palette = sns.color_palette(["#FF5733", "#33FF57"])  # Hex colors
sns.lineplot(data=student_data, x="Attendance", y="Exam_Score", hue="Gender", palette=custom_palette, errorbar="sd", style="Gender",  markers=True, dashes=False)


# Add title and labels
plt.title("How Attendance Affects Exam Scores")
plt.xlabel("Attendance (days)")
plt.ylabel("Exam Score")

# Add background
sns.set_style("whitegrid")

plt.show()
```
